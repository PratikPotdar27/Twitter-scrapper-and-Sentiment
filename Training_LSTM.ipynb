```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('sentiment.csv',encoding = "ISO-8859-1")
```

```python
data1=pd.DataFrame(columns=['target','tweets'])
```

```python
d1=data.iloc[:,0:1]
data1['target']=d1['0']
d2=data.iloc[:,5:10]
data1['tweets']=d2[0:1599999]
data1.head()
```

```python
new_data=data1[0:10000]
```

```python

###Drop Nan Values
new_data=new_data.dropna()
```

```python
x=new_data['tweets']
y=new_data['target']
```

```python
x.shape
```

```python
y.shape
```

```python

import tensorflow as tf
```

```python
tf.__version__
```

```python

from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
voc_size=5000
```

```python
messages=new_data.copy()
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
    print(i)
    review = re.sub('[^a-zA-Z]', ' ', messages['tweets'][i])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python

onehot_repr=[one_hot(words,voc_size)for words in corpus] 
onehot_repr
```

```python

sent_length=20
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
embedded_docs[0]
```

```python
embedding_vector_features=40
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
len(embedded_docs),y.shape
```

```python

import numpy as np
X_final=np.array(embedded_docs)
y_final=np.array(y)
```

```python

X_final.shape,y_final.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.33, random_state=42)
```

```python
from keras.callbacks import ModelCheckpoint
```

```python
import os
```

```python
checkpoint_path= "training_1/cp.ckpt"
checkpoint_dir= os.path.dirname(checkpoint_path)
```

```python
cp_callback=tf.keras.callbacks.ModelCheckpoint(filepath=checkpoint_path,save_weights_only=True,verbose=1)
```

```python
#checkpoint = ModelCheckpoint('model-{epoch:03d}.model',monitor='val_loss',verbose=0,save_best_only=True,mode='auto')
### Finally Training
p=model.fit(X_train,y_train,validation_data=(X_test,y_test),callbacks=[cp_callback],epochs=1,batch_size=64)
```

```python
from tensorflow.keras.layers import Dropout
## Creating model
embedding_vector_features=40
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(Dropout(0.3))
model.add(LSTM(100))
model.add(Dropout(0.3))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python

y_pred=model.predict_classes(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```

```python

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
